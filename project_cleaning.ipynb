```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.mstats import winsorize
```

```python
num_cols= df.select_dtypes(np.number)
num_cols
```

```python
for i in num_cols:
  plt.figure(figsize=(10,5))
  plt.boxplot(df[i])
  plt.title(i)
```

```python
columns_to_winsorize = ['Rain_fall_actual(mm)', 'Rain_fall_deviation(%)']
for crop in df['Crop'].unique():
    crop_mask = df['Crop'] == crop

    for col in columns_to_winsorize:
        crop_values = df.loc[crop_mask, col]
        df.loc[crop_mask, col] = winsorize(crop_values, limits=(0.01, 0.05))
```

```python
for i in columns_to_winsorize:
  plt.figure(figsize=(10,5))
  plt.boxplot(df[i])
  plt.title(i)
```

```python
import seaborn as sns
# Ensure 'Crop' is categorical
df['Crop'] = df['Crop'].astype('category')

# Loop through each crop and draw a boxplot
for crop in df['Crop'].unique():
    crop_data = df[df['Crop'] == crop]

    plt.figure(figsize=(8, 5))
    sns.boxplot(y='Yield', data=crop_data, color='skyblue')
    plt.title(f'Yield Distribution for {crop}', fontsize=14)
    plt.ylabel('Yield (kg/ha)', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# Example: threshold dictionary (you can adjust these based on your analysis)
thresholds = {
    'Banana': 100,    # kg/ha or tonne/ha depending on your unit
    'Pepper': 5,
    'Paddy': 10,
    'Tapioca': 100
}


# Loop through each crop and replace outliers
for crop, limit in thresholds.items():
    crop_mask = df['Crop'] == crop
    crop_data = df.loc[crop_mask, 'Yield']

    # Compute mean excluding outliers
    mean_without_outliers = crop_data[crop_data <= limit].mean()

    # Replace outlier yields
    df.loc[crop_mask & (df['Yield'] > limit), 'Yield'] = mean_without_outliers

    print(f"{crop}: Replaced yields > {limit} with mean {mean_without_outliers:.2f}")

# Optional: check before and after
print("\nOutlier treatment complete âœ…")
```

```python
# Function to winsorize within each crop
def winsorize_crop(data, col='grand_total_production', lower_quantile=0, upper_quantile=0.99):
    for crop in df['Crop'].unique():
        mask = df['Crop'] == crop
        lower = df.loc[mask, col].quantile(lower_quantile)
        upper = df.loc[mask, col].quantile(upper_quantile)
        df.loc[mask, col] = df.loc[mask, col].clip(lower, upper)
    return df

# Apply Winsorization
data = winsorize_crop(df, col='grand_total_production', lower_quantile=0, upper_quantile=0.99)
```
